1. Project Summary

This project looks at how stocks relate to each other and whether those relationships form meaningful groups.
The workflow is:

(1) compute log returns and correlations

(2) build a Minimum Spanning Tree (MST) to keep only the closest connections

(3) “cut” weak links to form clusters

(4) visualize the network and cluster behavior

For Deliverable 2, I added cluster summary statistics, correlation heatmaps, and a rolling-window version of the analysis to see how the structure changes over time.

2. Finance Background

Stocks often move together because they respond to similar market forces. Correlation is a standard way to measure how closely two stocks move. High correlation means they usually rise or fall together; low correlation means they behave more independently.

An MST is a tool that lets us reduce a fully connected graph into a simpler structure that keeps only the strongest relationships. Once the tree is built, removing weaker edges naturally forms groups of stocks that tend to move in similar ways.

Rolling windows help show whether these relationships are stable or if the market structure changes over different periods.
