1. Project Summary

This project explores how stocks are related to each other and whether those relationships form natural groups. I measure the similarity between stocks using basic statistical tools (correlation and covariance), build a Minimum Spanning Tree (MST) to keep only the closest connections, then “cut” the weak links to create clusters. Finally, I visualize the network structure and show how stocks inside the same cluster move together over time.

2. Finance Background

Stocks often move together because they’re affected by similar economic forces. Correlation is a standard way to measure how closely two stocks move. Higher correlation means two stocks tend to rise and fall together; lower correlation means they behave differently.
An MST lets us simplify a large network by keeping only the strongest links. Cutting weak connections helps us see groups of stocks that are closely related—similar to sectors, themes, or shared risk exposures.
